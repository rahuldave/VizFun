Make a scatter plot of Pennicilin (y-axis) vs Neomycin (x-axis) concentrations. Color code these via genus. Do you see something interesting concerning Diplococcus vs Streptococcus bacteria?

```python
# your code here
fig, ax = plt.subplots(figsize = (9,6))
for name, group in dfabio.groupby("genus"):
    ax.scatter(group.neomycin, group.penicillin, alpha=0.7, label=name)
ax.set_xlabel("neomycin")
ax.set_ylabel("pennicilin")
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = {'genus': ['A', 'B', 'C', 'A', 'B', 'C'],
        'x': [1, 2, 3, 4, 5, 6],
        'y': [10, 15, 20, 25, 30, 35]}

df = pd.DataFrame(data)

# Define a mapping of genus to color
genus_color_map = {'A': 'red', 'B': 'blue', 'C': 'green'}

# Map the 'genus' column to colors using the mapping
df['color'] = df['genus'].map(genus_color_map)

# Scatterplot with colors and legend
plt.figure(figsize=(8, 6))
scatter = plt.scatter(df['x'], df['y'], c=df['color'])

# Create a legend with one label per genus color
legend_labels = []
for genus, color in genus_color_map.items():
    legend_labels.append(plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, label=genus))

plt.legend(handles=legend_labels)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Scatterplot with Genus Labels and Legend')
plt.grid(True)
plt.show()
```
