```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
campdict = {
    "Excited":[19, 38],
    "Kind of Interested":[25, 30],
    "OK": [40, 14],
    "Not Great": [5, 6],
    "Bored": [11, 12]
}
states = ['Before',	'After']
```

```python
df = pd.DataFrame.from_dict(campdict).T
```

```python
df
```

```python
df.columns
```

```python
df.rename(columns={0:states[0], 1: states[1]})
```

```python
df = df.rename(columns={0:states[0], 1: states[1]})
```

```python
df
```

```python
df.Before.plot.pie()
```

```python
fig, axs = plt.subplots(nrows = 2, figsize=(5,8))
df.Before.plot.pie(ax=axs[0])
df.After.plot.pie(ax=axs[1])
```

```python
axs
```

```python
fig, axs = plt.subplots(ncols = 2, figsize=(12,5))
df.Before.plot.bar(ax=axs[0])
print(axs[0].get_ylim())
df.After.plot.bar(ax=axs[1])
axs[1].set_ylim(axs[0].get_ylim())
fig.tight_layout()
```
